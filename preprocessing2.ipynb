```python
def clean_price_column(price):
    if isinstance (price,str):
         price= price.replace(',','')
         price=price.strip()
         if price.isnumeric():
             price= int(price)
             return price
         if len(price)==0:
              return np.nan
    return price
```

```python
clean_price_column('1,999')
```

```python
df['cleaned_price']= df['price'].apply(clean_price_column)
df
```

```python
def clean_rating_column(rating):
    if isinstance(rating,str):
        rating= rating.split()[0]
        return float(rating)
    return rating
```

```python
df['cleaned_rating']=df['rating'].apply(clean_rating_column)
df
```

```python
df.sort_values(by='cleaned_price',ascending=False,inplace=True)
df
```

```python
df.drop(columns=['price','rating'], inplace=True)
df
```
