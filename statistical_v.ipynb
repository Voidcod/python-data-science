```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as mp
import numpy as np

```

```python
df= sns.load_dataset("flights")
df
```

```python
sns.relplot(x='passengers',y='month',data=df)
```

```python
sns.relplot(x='passengers',y='month',hue='year',data=df)
```

```python
b= sns.load_dataset("tips")
b
```

```python
sns.catplot(x='day',y='total_bill',data=b)
```

```python
sns.catplot(x='day',y='total_bill',kind='violin',data=b)
```

```python
sns.catplot(x='day',y='total_bill',kind='boxen',data=b)
```
