```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df= sns.load_dataset('iris')
df
```

```python
sns.histplot(df, x='sepal_width',bins=15,kde=True) # kernel density estimation plot
```

```python
fig,ax=plt.subplots(figsize=(8,4))
sns.histplot(df,x='sepal_length',bins=15,kde=True,ax=ax,color="red")
plt.savefig('histogram.png',dpi=300)
```

```python
sns.color_palette('rainbow')
```

```python
sns.set_palette('rainbow')
fig,ax=plt.subplots(figsize=(20,5),ncols=4)
sns.histplot(df,x='sepal_length',bins=15,kde=True,ax=ax[0],color="red")
sns.histplot(df,x='sepal_length',bins=15,kde=True,ax=ax[1],color="blue")
sns.histplot(df,x='sepal_length',bins=15,kde=True,ax=ax[2],color="green")
sns.histplot(df,x='sepal_length',bins=15,kde=True,ax=ax[3],color="yellow")
```
